The goal of this analysis is to evaluate how a person's health may be affected
if they consistently eat at a single fast-food restaurant.

As a reference point, recommended daily nutrient intake values for an average adult
were taken from World Health Organization (WHO) guidelines.

The analysis focuses on:
- Comparing daily nutrient intake to recommended norms
- Estimating potential health risks based on deviations from these norms
- Ranking restaurants by overall nutritional risk


 Recommended daily intake values (WHO-based, approximate)

```python
daily_norms = {
    'calories': 2000,
    'cal_fat' : 700,
    'sodium' : 2000,
    'calcium' : 1000,
    'total_fat': 70,
    'sugar': 50,
    'protein': 50,
    'sat_fat': 20,
    'fiber': 25,
    'total_carb': 260,
    'cholesterol': 300,
    'vit_a' : 100,
    'vit_c' : 100,
    'trans_fat' : 0
}
# Convert daily norms dictionary keys into a list for easier iteration

nutrients = list(daily_norms.keys())
```

Create a table describing potential health effects
associated with nutrient deficiency or excess,
as well as the approximate time scale of symptom onset.

```python
symptom_table = {
    "calories_from_fat": {
        "deficit": "Weight loss, fatigue, hormonal imbalance",
        "excess": "Obesity, insulin resistance"
    },
    "total_fat": {
        "deficit": "Hormonal imbalance, poor vitamin absorption",
        "excess": "Obesity, elevated cholesterol"
    },
    "saturated_fat": {
        "deficit": "-",
        "excess": "Increased risk of cardiovascular diseases, inflammation"
    },
    "trans_fat": {
        "deficit": "-",
        "excess": "Very high risk of cardiovascular diseases, inflammation, diabetes"
    },
    "cholesterol": {
        "deficit": "-",
        "excess": "Elevated LDL, risk of atherosclerosis"
    },
    "sodium": {
        "deficit": "Muscle cramps, low blood pressure, weakness",
        "excess": "Hypertension, cardiovascular diseases, kidney strain"
    },
    "carbohydrates": {
        "deficit": "Fatigue, poor concentration",
        "excess": "Weight gain, insulin spikes"
    },
    "fiber": {
        "deficit": "Constipation, poor gut microbiota, colon cancer risk",
        "excess": "Bloating, impaired nutrient absorption (rare)"
    },
    "sugar": {
        "deficit": "Hypoglycemia, lethargy",
        "excess": "Dental caries, obesity, type II diabetes"
    },
    "protein": {
        "deficit": "Muscle loss, weakness, anemia",
        "excess": "Kidney strain, osteoporosis risk (with low calcium intake)"
    },
    "vitamin_a": {
        "deficit": "Impaired immunity, skin and vision issues",
        "excess": "Toxicity: headaches, liver and bone damage"
    },
    "vitamin_c": {
        "deficit": "Fatigue, nervous system disorders, anemia, skin issues",
        "excess": "Usually excreted, but excess B6 may cause neuropathy"
    },
    "calcium": {
        "deficit": "Osteoporosis, muscle cramps",
        "excess": "Kidney stones, impaired absorption of other minerals"
    }
}



def get_symptom(nutrient, value):
    """
    Classifies potential health effects based on percentage deviation
    from the recommended daily intake.

    Parameters
    ----------
    nutrient : str
        Nutrient name.
    delta_percent : float
        Percentage deviation from the daily norm (%).

    Returns
    -------
    str
        Qualitative description of health risk severity.
    """
    if abs(value) <= 10:
        return "Negligible effect"
    elif value < -100:
        return "Metabolic and immune disorders: " + symptom_table.get(nutrient, {}).get("deficit", "-")
    elif value < -50:
        return "Direct harm within weeks or months: " + symptom_table.get(nutrient, {}).get("deficit", "-")
    elif value < -25:
        return "Noticeable symptoms within months: " + symptom_table.get(nutrient, {}).get("deficit", "-")
    elif value < -10:
        return "Mild long-term symptoms: " + symptom_table.get(nutrient, {}).get("deficit", "-")
    elif value > 100:
        return "High risk of pathological conditions: " + symptom_table.get(nutrient, {}).get("excess", "-")
    elif value > 50:
        return "Direct harm within weeks or months: " + symptom_table.get(nutrient, {}).get("excess", "-")
    elif value > 25:
        return "Noticeable symptoms within months: " + symptom_table.get(nutrient, {}).get("excess", "-")
    elif value > 10:
        return "Mild long-term symptoms: " + symptom_table.get(nutrient, {}).get("excess", "-")
    else:
        return "Negligible effect"
```


Functions for evaluating health risks associated with dietary patterns

```python
from collections import defaultdict
def risk_level_key(symptom):
    #     get_symptom
    if symptom.startswith("High risk"):
        return "High risk of pathological conditions"
    elif symptom.startswith("Direct harm"):
        return "Direct harm within weeks or months"
    elif symptom.startswith("Noticeable symptoms"):
        return "Noticeable symptoms within months"
    elif symptom.startswith("Mild symptoms"):
        return "Mild long-term symptoms"
    else:
        return None


# Generate a final report for each restaurant

def make_risk_report(row):
    risks = defaultdict(list) # key  risk level, value  list of health effects


    for nutrient, symptom in row.items():
        key = risk_level_key(symptom)
        if not key:
            continue

        # Extract the health effect description after the colon ": "
        # If no colon is present, use the full text as a fallback

        desc = symptom.split(": ", 1)[-1] if ": " in symptom else symptom

        # Skip empty or placeholder descriptions ("-")

        if desc.strip() in ("-", ""):
            continue
        # Skip adding descriptions for negligible effects
        if key == "Negligible effect":
            continue

        # Add the description to the corresponding risk level

        risks[key].append(desc.strip())

    if not risks:
        return "    ."

    # Build a textual report grouped by risk severity levels

    result_lines = []
    # Sort risk levels by severity

    priority_order = [
        "High risk of pathological conditions",
        "Direct harm within weeks or months",
        "Noticeable symptoms within months",
        "Mild long-term symptoms",
        "Negligible effect"
    ]

    for level in priority_order:
        if level in risks:
            # Remove duplicates and sort

            unique_desc = sorted(set(risks[level]))
            result_lines.append(f"{level}:")
            for d in unique_desc:
                result_lines.append(f"  - {d}")
            result_lines.append("") # Empty line between risk groups


    return "\n".join(result_lines).strip()
```

```python
# Weights assigned to each risk level

risk_weights = {
    "High risk of pathological conditions":4,
    "Direct harm within weeks or months":3,
    "Noticeable symptoms within months":2,
    "Mild long-term symptoms": 1,
    "Negligible effect" :0,
}

def calculate_risk_score(row):
    risks = defaultdict(int)
    for nutrient, symptom in row.items():
        key = risk_level_key(symptom)
        if not key:
            continue
        # Add +1 count to this risk category
        risks[key] += 1

    # Calculate the total weighted risk score
    total_score = 0
    for level, count in risks.items():
        weight = risk_weights.get(level, 0)
        total_score += weight * count

    return total_score
```

```python

def risk_level_key(symptom_text):
    if not isinstance(symptom_text, str):
        return " "
    if symptom_text.startswith("High risk of pathological conditions"):
        return "High risk of pathological conditions"
    elif symptom_text.startswith("Direct harm within weeks or months"):
        return "Direct harm within weeks or months"
    elif symptom_text.startswith("Noticeable symptoms within months"):
        return "Noticeable symptoms within months"
    elif symptom_text.startswith("Mild long-term symptoms"):
        return "Mild long-term symptoms"
    else:
        return "Negligible effect"

risk_levels = [
    "High risk of pathological conditions",
    "Direct harm within weeks or months",
    "Noticeable symptoms within months",
    "Mild long-term symptoms",
    "Negligible effect"
]
```


#Model 1: Satiety-based approach

This model estimates daily nutrient intake assuming that a person eats
until a target level of satiety is reached.

Satiety is estimated using a weighted formula based on protein, fiber,
fat, and sugar content.




 Satiety score formula:
 Satiety = (2  protein) + (1.5  fiber) + (1  fat)  (0.5  sugar)
 Higher weights are assigned to protein and fiber due to their stronger
 effect on perceived fullness, as supported by nutrition research.

```python
df['satiety'] = (
    2 * df['protein'] +
    1.5 * df['fiber'] +
    df['total_fat'] -
    0.5 * df['sugar']
)
```

 Median nutrient values per restaurant are used
 as a representative "typical" meal.

```python

median_dish = df.groupby('restaurant')[nutrients].median()
"""
Trans fats require special handling.
Using percentage deviation with a zero reference value leads to NaN or infinity,
therefore the mean value is used instead of the median.
"""

mean_trans = df.groupby('restaurant')['trans_fat'].mean()
median_dish['trans_fat'] = mean_trans
median_dish
```


According to the satiety formula, a value above 60
is considered indicative of a filling meal.



```python
median_satiety = df.groupby('restaurant')['satiety'].median()
median_satiety
```

A total satiety score of 210 is used as a proxy for three meals per day,
assuming an average satiety score of 70 per meal.


```python

meals_per_day = (210 / median_satiety).round()
meals_per_day
```

```python

adjusted_nutrients = median_dish.multiply(meals_per_day, axis=0)

delta_adjusted = adjusted_nutrients.subtract(pd.Series(daily_norms))
daily_norms_df = pd.DataFrame([daily_norms] * len(adjusted_nutrients),
                              index=adjusted_nutrients.index)
```

Approximate daily nutrient intake based on the satiety-based model

```python

nutrients_per_day = median_dish.multiply(meals_per_day, axis=0)
nutrients_per_day
```

```python

# Add daily recommended values for comparison with restaurants
nutrients_per_day.loc['Daily Norm'] = daily_norms
```

```python

# Build the bar chart
for nutrient in nutrients:
    plt.figure(figsize=(10, 4))
    nutrients_per_day[nutrient].plot(kind='bar', color='skyblue', edgecolor='black')
    plt.axhline(y=daily_norms[nutrient], color='red', linestyle='--', label='Daily Norm')
    plt.title(f'{nutrient.capitalize()} per Day by Restaurant')
    plt.ylabel(f'{nutrient} amount')
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.grid(True, axis='y')
    plt.show()
```

```python

# Relative deviation (percentage): (intake  recommended) / recommended  100

delta_percent_meals = ((adjusted_nutrients - daily_norms_df) / daily_norms_df) * 100

# Trans fats are handled separately:
# 0 mg is optimal, while 2 mg is considered the upper acceptable limit
def transform_trans_fat(value, norm=2.0):
    if value == 0:
        return 0  # optimal
    elif 0 < value <= norm:
        return (value / norm) * 100  # mild or moderate deviation
    else:
        return ((value - norm) / norm) * 100 + 100  # severe excess

# Apply the transformation to the column
delta_percent_meals['trans_fat'] = adjusted_nutrients['trans_fat'].apply(transform_trans_fat)
# Build a heatmap of percentage deviations
plt.figure(figsize=(14, 8))
sns.heatmap(delta_percent_meals, cmap='coolwarm', center=0, annot=True, fmt=".0f")
plt.title("Deviation from recommended intake (%) by nutrients and restaurants")
plt.xlabel("Nutrients")
plt.ylabel("Restaurants")
plt.tight_layout()
plt.show()
# Function to classify nutritional risk based on percentage deviation

def classify_risk(value):
    abs_val = abs(value)
    if abs_val <= 10:
        return "Normal"
    elif abs_val <= 25:
        return "Mild"
    elif abs_val <= 50:
        return "Moderate"
    elif abs_val <= 100:
        return "Severe"
    else:
        return "Severe deficiency" if value < 0 else "Toxicity risk"
risk_levels_classify_m = delta_percent_meals.applymap(classify_risk)

# Normalize values to enhance contrast in the heatmap
risk_scale = {
    "Normal": 0,
    "Mild": 1,
    "Moderate": 2,
    "Severe": 3,
    "Severe deficiency": 4,
    "Toxicity risk": 5
}
numeric_risks = risk_levels_classify_m.replace(risk_scale)
# Build a heatmap showing risk severity levels
plt.figure(figsize=(14, 8))
sns.heatmap(numeric_risks, cmap="YlOrRd", annot=False, fmt='', cbar_kws={"label": "Level of risk"})
plt.title("Risk categories by nutrient")
plt.xlabel("Nutrients")
plt.ylabel("Restaurants")
plt.tight_layout()
plt.show()
```

The heatmaps indicate that although the diet is not immediately catastrophic,
it consistently leads to harmful nutritional imbalances.

```python
symptom_df_meals = delta_percent_meals.copy()

for nutrient in delta_percent_meals.columns:
    symptom_df_meals[nutrient] = delta_percent_meals[nutrient].apply(lambda val: get_symptom(nutrient, val))
```

```python
# Output results for all restaurants
for restaurant, row in symptom_df_meals.iterrows():
    print(f"{restaurant}\n")
    print(make_risk_report(row))
    print("\n" + "-"*50 + "\n")
```

Quantifying health impact by assigning equal weight to all symptoms
and considering only the speed of their potential onset

```python
# Create a new column with risk scores
symptom_df_meals['risk_score'] = symptom_df_meals.apply(calculate_risk_score, axis=1)

# Sort by descending risk level
sorted_risks = symptom_df_meals['risk_score'].sort_values(ascending=False)

print("Restaurant ranking by overall health risk (higher = worse):\n")
for restaurant, score in sorted_risks.items():
    print(f"{restaurant}: {score}")
```

```python

# Count the number of symptoms per risk level for each restaurant
risk_counts_per_restaurant = {}

for idx, row in symptom_df_meals.iterrows():
    counts = defaultdict(int)
    for nutrient, symptom in row.items():
        level = risk_level_key(symptom)
        counts[level] += 1
    risk_counts_per_restaurant[idx] = counts

risk_counts_df = pd.DataFrame(risk_counts_per_restaurant).fillna(0).T[risk_levels]

# Build a horizontal stacked bar chart
fig, ax = plt.subplots(figsize=(16, 8))

left = pd.Series([0]*len(risk_counts_df), index=risk_counts_df.index)
colors = {
    "High risk of pathological conditions": "#d62728",
    "Direct harm within weeks or months": "#ff7f0e",
    "Noticeable symptoms within months": "#2ca02c",
    "Mild long-term symptoms": "#1f77b4",
    "Negligible effect": "#7f7f7f",
}


for level in risk_levels:
    ax.barh(risk_counts_df.index, risk_counts_df[level], left=left, label=level, color=colors[level])
    left += risk_counts_df[level]

ax.set_xlabel("Number of risk-related symptoms")
ax.set_title("Risk distribution by symptom severity across restaurants")
ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, 1.15),
    ncol=5,
    fancybox=True,
    shadow=True
)
plt.subplots_adjust(top=0.85)
plt.show()
```

Interpretation of Model 1 results

This model suggests that, for most nutrients, deviations from recommended
daily intake are moderate. As a result, severe symptoms are unlikely to appear
immediately, but mild to moderate health effects may develop over weeks or months
of sustained dietary behavior.

A key limitation of this approach is the presence of a significant energy deficit
under full satiety conditions. Even when a person eats until feeling full,
the total caloric intake often remains below daily energy requirements.
This would likely lead to chronic fatigue and reduced physical and cognitive performance.

In real life, such an energy deficit would probably be compensated by consuming
additional meals or snacks that are energy-dense but not necessarily satiating,
potentially worsening nutritional balance.

#Model 2: Energy-balanced scenario

To address this limitation, a second model is introduced.
In this model, the same representative dishes are used, but the total quantity
of food is scaled so that daily caloric intake exactly meets the recommended
energy requirement.
This scenario serves as a baseline comparison, isolating nutritional imbalances
that remain even when energy needs are fully satisfied.

Limitations of the analysis:

- The model assumes that a person repeatedly eats the same "median" meal,
  which is not realistic in real life.
- Individual differences (age, sex, activity level, health conditions)
  are not taken into account.
- Satiety is modeled using a simplified heuristic formula.
- The second model is intentionally simplistic and serves as a baseline scenario.

Despite these limitations, the analysis highlights systematic nutritional
imbalances present in fast-food diets.
